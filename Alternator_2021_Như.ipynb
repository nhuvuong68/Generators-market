```python
# Read file
data = pd.read_excel('Bao cao 8501 (T1 - T11.2021).xlsx', header = 0, sheet_name = 'Bao cao nhap khau_working')

print(data.shape) # Number of rows & columns

# Uppercase to standardize
data['IMPORTER'] = data['IMPORTER'].str.upper() 
data['DETAILED COMMODITY'] = data['DETAILED COMMODITY'].str.upper()

data.head()
```

```python
# Create column unit price USD
data['UNIT PRICE USD'] = np.divide (data['AMOUNT (USD)'], data['QUANTITY'])
data['UNIT PRICE USD'] = data['UNIT PRICE USD'].round(2)
print(data['UNIT PRICE USD'])
```

```python
# Filter new/used


data.loc[data['DETAILED COMMODITY'].str.contains('100%|MỚI 100 %|HÀNG MỚI|MỚI 10%'), 'CHECK NEW'] = 'New'
data.loc[data['DETAILED COMMODITY'].str.contains('ĐÃ QUA SỬ DỤNG|NĂM 2014|NĂM 2016|ĐQSD|(DQSD)|THANH LÝ|ĐÃ QUA SD|ĐÃ QUA S|SẢN XUẤT 2014|ĐÃ SỬ DỤNG|QUA SỬ DỤNG|ĐÃ QUA SỬ DUNG|ĐÃ QUA SỬ DỤNG|SẢN XUẤT 2012-2013|SẢN XUẤT 2012|ĐÃ QUA  SỬ DỤNG|SẢN XUẤT NĂM 2013'), 'CHECK NEW'] = 'Used'
data['CHECK NEW'] = data['CHECK NEW'].fillna('N/A')
# Check data to collect data list
# data.loc[data['CHECK NEW'] == 'N/A' ]['DETAILED COMMODITY'][14001:15000,].unique()
data['CHECK NEW'].value_counts()
```
